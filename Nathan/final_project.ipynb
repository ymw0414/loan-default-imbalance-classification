# Introduction

This project focuses on predicting whether a client will default on a loan using the Home Credit Default Risk dataset from Kaggle. The objective is to build a model that estimates the probability that a borrower will fail to repay a loan, based on a wide range of financial and personal characteristics. The dataset is provided by Home Credit, a consumer finance company that aims to expand access to credit for individuals with limited or no traditional credit history. This makes the problem especially relevant, as standard credit scoring methods may not perform well for such populations.

The data includes detailed information about each loan application, such as demographic variables, income, employment history, credit history, and loan characteristics. In addition, the full competition provides access to several related tables that capture past credit behavior and previous applications. In this project, the main application dataset is used, which already contains a large number of features and reflects common real world challenges such as missing values, categorical variables, and a strong class imbalance, since most clients do not default.

The motivation for this task lies in the importance of accurate credit risk assessment. Better predictions of default risk allow lenders to reduce financial losses and make more informed lending decisions, while also expanding access to credit in a responsible way. From a modeling perspective, several machine learning approaches are applied and compared, including Logistic Regression, Random Forest, Histogram Gradient Boosting, and a neural network. These models provide a range of complexity, from simple linear methods to more flexible nonlinear approaches, allowing for a comprehensive evaluation of predictive performance on this dataset.

# Hyperparameter Tuning

#### Creating the dictionary of the models

```python
# Import the models we want to compare
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.experimental import enable_hist_gradient_boosting
from sklearn.ensemble import HistGradientBoostingClassifier

# Create a dictionary of models
models = {
    'lr': LogisticRegression(class_weight='balanced', random_state=random_seed),
    'rfc': RandomForestClassifier(class_weight='balanced', random_state=random_seed),
    'hgbc': HistGradientBoostingClassifier(random_state=random_seed)
}
```

#### Creating the dictionary of the pipelines

```python
# Wrap each model in a pipeline (keeps the code flexible and consistent)
from sklearn.pipeline import Pipeline

pipes = {}

for acronym, model in models.items():
    pipes[acronym] = Pipeline([
        ('model', model)
    ])
```

#### Getting the predefined split cross-validator

```python
# Get the:
# feature matrix and target velctor in the combined training and validation data
# target vector in the combined training and validation data
# PredefinedSplit
# See the implementation in pmlm_utilities.ipynb
X_train_val, y_train_val, ps = get_train_val_ps(X_train, y_train, X_val, y_val)

# note: this way, GridSearchCV tunes hyperparameters using fixed train/val split,
# instead of doing its own random cross-validation folds.
```

### GridSearchCV

#### Creating the dictionary of the parameter grids

```python
# For each model, define the hyperparameter values that GridSearchCV will try
param_grids = {}


# Logistic Regression
# tol: stopping tolerance
# C: inverse of regularization strength
tol_grid = [1e-5, 1e-4, 1e-3]
C_grid = [0.01, 0.1, 1, 10]

param_grids['lr'] = [{
    'model__tol': tol_grid,
    'model__C': C_grid
}]


# Random Forest
# n_estimators: number of trees
# min_samples_split: minimum samples needed to split a node
# min_samples_leaf: minimum samples in a leaf
n_estimators_grid = [100, 200]
min_samples_split_grid = [2, 20, 100]
min_samples_leaf_grid = [1, 20, 100]

param_grids['rfc'] = [{
    'model__n_estimators': n_estimators_grid,
    'model__min_samples_split': min_samples_split_grid,
    'model__min_samples_leaf': min_samples_leaf_grid
}]


# HistGradientBoostingClassifier
# learning_rate: step size of boosting updates
# max_iter: number of boosting iterations
# min_samples_leaf: minimum samples in a leaf
learning_rate_grid = [0.01, 0.05, 0.1]
max_iter_grid = [100, 200]
min_samples_leaf_grid = [20, 100]

param_grids['hgbc'] = [{
    'model__learning_rate': learning_rate_grid,
    'model__max_iter': max_iter_grid,
    'model__min_samples_leaf': min_samples_leaf_grid
}]
```

#### Creating the directory for the cv results produced by GridSearchCV

```python
import os
directory = os.path.dirname(abspath_curr + '/result/home_credit/cv_results/GridSearchCV/')
if not os.path.exists(directory):
    os.makedirs(directory)
```

#### Tuning the hyperparameters

```python
#-------------------------------------------------------------------------------
# Run GridSearchCV for each model
from sklearn.model_selection import GridSearchCV

# This list will store, for each model: [best_score_, best_params_, best_estimator_]
best_score_params_estimator_gs = []

for acronym in pipes.keys():

#-------------------------------------------------------------------------------
# Set up GridSearchCV
    gs = GridSearchCV(
        estimator=pipes[acronym],
        param_grid=param_grids[acronym],
        scoring='roc_auc', # works good with imbalanced data & used often for credit default
        n_jobs=2,
        cv=ps,
        return_train_score=True
    )

# Fit the model on the combined training+validation data, using the predefined split object
    gs = gs.fit(X_train_val, y_train_val)

# Save the best score, best parameters, and best fitted estimator
    best_score_params_estimator_gs.append([
        gs.best_score_,
        gs.best_params_,
        gs.best_estimator_
    ])

# Convert cv_results_ to a DataFrame and sort it
    cv_results = pd.DataFrame.from_dict(gs.cv_results_).sort_values(
        by=['rank_test_score', 'std_test_score']
    )

# Put the most important columns first
    important_columns = [
        'rank_test_score',
        'mean_test_score',
        'std_test_score',
        'mean_train_score',
        'std_train_score',
        'mean_fit_time',
        'std_fit_time',
        'mean_score_time',
        'std_score_time'
    ]

    cv_results = cv_results[
        important_columns + sorted(list(set(cv_results.columns) - set(important_columns)))
    ]

# Save the cross-validation results for this model
    cv_results.to_csv(
        path_or_buf=abspath_curr + '/result/home_credit/cv_results/GridSearchCV/' + acronym + '.csv',
        index=False
    )

#-------------------------------------------------------------------------------
# Sort the models by best validation score, highest first
best_score_params_estimator_gs = sorted(
    best_score_params_estimator_gs,
    key=lambda x: x[0],
    reverse=True
)

# Display model comparison table
pd.DataFrame(
    best_score_params_estimator_gs,
    columns=['best_score', 'best_param', 'best_estimator']
)
```
